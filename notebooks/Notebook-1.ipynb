```python
# find rating avges for every book in the books_df slice

def avge_rating_for_one_book_across_slices(book_ISBN):
    #extract ratings for our book from 1st ratings slice
    first_ratings = list_ratings_df[0]
    book_ratings = pd.DataFrame()
    book_ratings = first_ratings.loc[first_ratings['ISBN'] == book_ISBN,:]
    
    #add ratings for our book across all slices of ratings_df to book_ratings
    for index in range(len(list_ratings_df)):
        our_slice = list_ratings_df[index]
        slice_ratings = our_slice.loc[our_slice['ISBN'] == book_ISBN,:]
        if (index > 0) : book_ratings = book_ratings.append(slice_ratings, ignore_index=True)

    # do the sum of all those ratings
    book_avg_rating_sum = 0.0
    for index, rating in book_ratings.iterrows():
        book_avg_rating_sum += int(rating['Book-Rating'])

    # divide by nb of ratings for this book
    book_avg_rating = 0.0
    book_avg_rating = book_avg_rating_sum / len(book_ratings)
    
    return book_avg_rating
```
